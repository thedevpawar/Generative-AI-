## Experiment 6: Implement the evaluation metrics using appropriate libraries or frameworks.

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt

#  Step 1: Updated Data
y_true = np.array([0,1,2,0,1,2,0,1,2,0,1,2,1,0,2,1,0,2,1,2,0,1,2,1,0])
y_pred = np.array([0,1,2,0,2,1,0,1,2,0,1,2,1,0,2,1,0,2,0,2,0,1,1,1,0])

#  Step 2: Overall Metrics
accuracy = accuracy_score(y_true, y_pred)
precision_macro = precision_score(y_true, y_pred, average='macro')
recall_macro = recall_score(y_true, y_pred, average='macro')
f1_macro = f1_score(y_true, y_pred, average='macro')

print(" - Overall Metrics:")
print(f"Accuracy        : {accuracy:.4f}")
print(f"Precision(Macro): {precision_macro:.4f}")
print(f"Recall(Macro)   : {recall_macro:.4f}")
print(f"F1 Score(Macro) : {f1_macro:.4f}")

#  Step 3: Class-wise Metrics
precision_class = precision_score(y_true, y_pred, average=None)
recall_class = recall_score(y_true, y_pred, average=None)
f1_class = f1_score(y_true, y_pred, average=None)

print("\n - Class-wise Metrics:")
for i in range(len(precision_class)):
    print(f"Class {i} -> Precision: {precision_class[i]:.3f}, Recall: {recall_class[i]:.3f}, F1: {f1_class[i]:.3f}")

#  Step 4: Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Step 5: Visualization
plt.figure(figsize=(14, 4))

# Graph 1: Overall Metrics
plt.subplot(1, 3, 1)
metrics = ['Accuracy', 'Precision', 'Recall', 'F1']
values = [accuracy, precision_macro, recall_macro, f1_macro]
plt.bar(metrics, values)
plt.ylim(0, 1)
plt.title("Overall Metrics")

# Graph 2: Confusion Matrix
plt.subplot(1, 3, 2)
plt.imshow(cm)
plt.colorbar()
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

# Graph 3: Class-wise Metrics
plt.subplot(1, 3, 3)
x = np.arange(len(precision_class))
width = 0.25

plt.bar(x - width, precision_class, width, label='Precision')
plt.bar(x, recall_class, width, label='Recall')
plt.bar(x + width, f1_class, width, label='F1 Score')

plt.xticks(x, [f"Class {i}" for i in range(len(x))])
plt.ylim(0, 1)
plt.title("Class-wise Metrics")
plt.legend()

plt.tight_layout()
plt.show()
```
